```python
import numpy as np # importamos la libreria numpy para poder utilizar el np.nan
def cob_nuloanan(df):
    df.replace("Nulo", np.nan, inplace=True) #aqui quitamos los valores nulos
```

```python
cob_nuloanan(df)
```

```python
(df == "Nulo").sum().sum() #verificamos que no ya no hay valores nulos
```

```python
def camb_valores(df): # esta funcion nos ayuda a cmabiar los valores de las columnas a numeros , con el errors='coerce', y manejar error con nan
    columnas = ["PRECIP","EVAP","TMAX","TMIN"] 
    for col in columnas:
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
camb_valores(df)
```

```python
df.info() # entendemos como estan organizados los datos y si hay valores faltantes
```

```python
def camb_fecha(df):
    df["FECHA"] = pd.to_datetime(df["FECHA"], dayfirst=True, errors='coerce') # se ponen los valores como nulos que esten  mal escritos 
    #o no sean feachas validas
```

```python
camb_fecha(df)
```
